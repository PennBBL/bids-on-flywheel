# A Quick Fix

Noticed that FNDM1 and FNDM2 were switched session labels. FNDM1 should have ITC tasks, while FNDM2 should have Card/Face/Perfusion. Fix is below

```python
# modules
import flywheel
import pandas as pd
# add the script to the path
import sys
import os
sys.path.append(os.path.abspath("/home/ttapera/bids-on-flywheel/flywheel_bids_tools"))
import query_bids
import upload_bids
from tqdm import tqdm
import math
```

```python
# query
fw = flywheel.Client()
result = query_bids.query_fw("Reward2018", fw)
```

```python
# clean query to get a table
view = fw.View(columns='subject')
subject_df = fw.read_view_dataframe(view, result.id)
sessions = []
view = fw.View(columns='acquisition')
pbar = tqdm(total=100)

for ind, row in subject_df.iterrows():
    session = fw.read_view_dataframe(view, row["subject.id"])
    if(session.shape[0] > 0):
        sessions.append(session)
    pbar.update(10)
pbar.close()
```

```python
acquisitions = pd.concat(sessions,ignore_index=True)
```

```python
acquisitions.head()
```

Here are the fndm2 acquisitions; you can see the acquisition labels are incorrect

```python
acquisitions.loc[acquisitions['session.label'] == "fndm2",]
```

We assign indeces and swap the session labels for them:

```python
ind2 = acquisitions.loc[acquisitions['session.label'] == "fndm2",].index
ind1 = acquisitions.loc[acquisitions['session.label'] == "fndm1",].index
```

```python
acquisitions.loc[ind2, 'session.label'] = "fndm1"
acquisitions.loc[ind1, 'session.label'] = "fndm2"
```

```python
acquisitions.loc[acquisitions['session.label'] == "fndm2",]
```

Now that the table is corrected, we upload the changes:

```python
fndm2 = acquisitions.loc[acquisitions['session.label'] == "fndm2",].reset_index()
```

```python
# get the session object
ses = fw.get(fndm2.loc[0,'session.id'])
```

```python
# update it's label
ses.update({"label": fndm2.loc[0,'session.label']})
```

```python
#reload to confirm
ses = ses.reload()
```

```python
ses.label
```

Here's a loop to automate the process:

```python
fndm2 = acquisitions.loc[acquisitions['session.label'] == "fndm2",].reset_index()
fndm2 = fndm2.drop_duplicates('session.id').reset_index()
fndm1 = acquisitions.loc[acquisitions['session.label'] == "fndm1",].reset_index()
fndm1 = fndm1.drop_duplicates('session.id').reset_index()

for ind, row in fndm2.iterrows():
    

    ses = fw.get(row['session.id'])
    ses.update({"label": row['session.label']})

for ind, row in fndm1.iterrows():
    
    ses = fw.get(row['session.id'])
    ses.update({"label": row['session.label']})
```

We re-query flywheel for an updated view:

```python
result = query_bids.query_fw("Reward2018", fw)

view = fw.View(columns='subject')
subject_df = fw.read_view_dataframe(view, result.id)
sessions = []
view = fw.View(columns='acquisition')
pbar = tqdm(total=100)

for ind, row in subject_df.iterrows():
    session = fw.read_view_dataframe(view, row["subject.id"])
    if(session.shape[0] > 0):
        sessions.append(session)
    pbar.update(10)
pbar.close()

acquisitions = pd.concat(sessions,ignore_index=True)
```

Finally, we see that FNDM1 has card and face tasks...
