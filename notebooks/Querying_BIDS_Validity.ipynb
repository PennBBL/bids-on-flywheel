# Querying Bids Validity

This notebook quickly demonstrates how to query flywheel for a project's BIDS validity using a custom script, which you can view [here](./scripts/query_bids_validity.py).

```python
import flywheel
import pandas
# add the script to the path
import sys
import os
sys.path.append(os.path.abspath("/home/ttapera/bids-on-flywheel/scripts"))
import query_bids_validity
```

First, initialise a flywheel client:

```python
fw = flywheel.Client()
```

Then, query your project, which returns a dataframe:

```python
%%time
df = query_bids_validity.query_bids_validity("Reward2018", fw, verbose=False)
```
