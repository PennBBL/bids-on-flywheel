# A Quick Fix Example 2

Noticed that B0 map dicoms are not classified and not run through dcm2nix. Fix is below

```python
# modules
import flywheel
import pandas as pd
# add the script to the path
import sys
import os
sys.path.append(os.path.abspath("/home/ttapera/bids-on-flywheel/flywheel_bids_tools"))
import query_bids
import upload_bids
from tqdm import tqdm
import math
```

```python
# query
fw = flywheel.Client()
result = query_bids.query_fw("Reward2018", fw)
```

```python
# clean query to get a table
view = fw.View(columns='subject')
subject_df = fw.read_view_dataframe(view, result.id)
sessions = []
view = fw.View(columns='acquisition')
pbar = tqdm(total=100)

for ind, row in subject_df.iterrows():
    session = fw.read_view_dataframe(view, row["subject.id"])
    if(session.shape[0] > 0):
        sessions.append(session)
    pbar.update(10)
pbar.close()
```

```python
acquisitions = pd.concat(sessions,ignore_index=True)
```

```python
b0maps = acquisitions.loc[acquisitions['acquisition.label'].str.contains("B0map"),].reset_index()
b0maps
```

We can see the BIDS classification hasn't been done correctly for these:

```python
acq = fw.get(b0maps.loc[2,'acquisition.id'])
```

We can assign a classification dictionary to the dicom of the form `'classification': {'x': ['y']}`

```python
f = acq.files[0]
```

```python
f.update({'modality': 'MR'})
f.update_classification({'Intent': ['Fieldmap'], 'Measurement': ['B0']})
```

Now we can loop through the full dataset to fix:

```python
classification = {'Intent': ['Fieldmap'], 'Measurement': ['B0']}

for ind, row in b0maps.iterrows():
    
    acq = fw.get(row['acquisition.id'])

    for f in acq.files:
        if f.type == "dicom":
            if f.modality is None:
                f.update({'modality': 'MR'})
            f.update_classification(classification)
```
