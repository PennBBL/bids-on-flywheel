In order to automate this process, we have to have a few steps:

1. Knowing the inputs and configurations available
2. Automating the creation of inputs and configurations
3. Running en masse

Let's list the inputs for all of the gears:

```python
import json
def collect_gear_config(gear_id):
    '''
    Collects the gear's configuration and inputs
    '''
    gear = fw.get_gear(gear_id)
    name = gear['gear']['name']
    label = gear['gear']['label']
    description = gear['gear']['description']
    inputs = nested_to_record(gear.gear.inputs)
    config = nested_to_record(gear.get_default_config())
    return({'name': name, 'inputs': inputs, 'config': config, 'label': label, 'description': description})
```

```python
random_g = df['id'].sample(1).values[0]
data = collect_gear_config(random_g)
print(json.dumps(data, indent=4))
```
